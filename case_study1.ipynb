# Find the top 5 cars based on the sales price

```python
data1=data.nlargest(5, ['Sale Price'])
```

```python
data1
```

```python
plt.bar(data1['Model'], data1['Sale Price'])
plt.xlabel('Cars')
plt.ylabel('Sale Price')
```

The top 5 cars based on sales price are Jetta III, A6, LX, MX-6 and ES

# The least 5 cars based on their Resell price

```python
data2=data.nsmallest(5, ['Resell Price'])
```
