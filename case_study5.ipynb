# ● Check whether all the zones generate the same amount of sales.

We should conduct ANOVA(Analysis of Variance) to test this.

Null Hypothesis(H_0):All Zones generate the same amount of sales

Alternative Hypothesis(H_a):All Zones doesn't generate the same amount of sales.

Significance level=5%

Since significance level is 5%, if p-value is less than 0.05, we reject null hypothesis

```python
stat.f_oneway(data['Zone - A'],data['Zone - B'],data['Zone - C'],data['Zone - D'])
```

Since p-value is less than 0.05 , we reject null hypothesis and conclude that the sales generated by each Zones are not same.
